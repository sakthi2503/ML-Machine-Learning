```python
import pandas as pb
```

```python
dataset=pb.read_csv("insurance_pre.csv")
```

```python
dataset=pb.get_dummies(dataset,drop_first=True)
```

```python
dataset.columns
```

```python
independant=dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
dependant=dataset[["charges"]]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(independant,dependant,test_size=0.30, random_state=0)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor(splitter= "random",criterion="poisson")
regressor.fit(X_train,Y_train)
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
tree.plot_tree(regressor)
plt.show()
```

```python
Y_predict=regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2_score=r2_score(Y_test,Y_predict)
r2_score
```
